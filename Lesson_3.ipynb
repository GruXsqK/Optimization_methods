__1. Сталепрокатное предприятие выпускает стандартные стальные листы размерами 1500×6000 мм. Ежемесячный объем заказов на листы нестандартных размеров приведён ниже: 500×600 — 3000, 1000×400 — 2000, 1500×1000 — 1500 Необходимо составить ЦФ для нахождения наименее затратного раскроя листов.__

__2. и составить алгоритм__

Пусть у нас имеется m вариантов раскроя, т.е. известны величины $a_{ij} (i= \overline{1,m}, j=\overline{1,3})$, определяющие количество единиц j-x изделий при i-м способе раскроя листа и обрезки $a_i$.

Пусть $x_i$ - количество листов, раскраиваемые i-м вариантом $(i=\overline{1,m})$

Тогда количество листов 500×600 равно:

$a_{11}x_1+...+a_{i1}x_i+...+a_{m1}x_m = 3000$

И аналогично для 1000×400 и 1500×1000 листов соответственно:

$a_{12}x_1+...+a_{i2}x_i+...+a_{m2}x_m = 2000$<br>
$a_{13}x_1+...+a_{i3}x_i+...+a_{m3}x_m = 1500$

Целевая функция:

$\sum_{i=1}^{m}x_ia_i \rightarrow min$

Итого:

$\sum_{i=1}^{m}x_ia_i \rightarrow min$<br>
$a_{1j}x_1+...+a_{ij}x_i+...+a_{mj}x_m = k_j$<br>

где $k_j$ - необходимое количество листов нестандартных размеров

```python
# поскольку листы 500×600 и 1500×1000 можно раскроить без остатков, то можно упростить алгоритм

st_sheet = (1500, 6000)
ord_sheets = [(500, 600), (1000, 400), (1500, 1000)]
n_ord_sheets = [3000, 2000, 1500]

for sheet, n_sheets in zip(ord_sheets, n_ord_sheets):
    # количество нестандартных листов при лучшем расположении их на стандартном листе
    best_cutting = max((st_sheet[0] // sheet[0]) * (st_sheet[1] // sheet[1]), 
                       (st_sheet[0] // sheet[1]) * (st_sheet[1] // sheet[0]))
    # округление до целого числа листов
    if n_sheets % best_cutting:
        n_st_sheets = int(round(n_sheets / best_cutting) + 1)
    else:
        n_st_sheets = int(n_sheets // best_cutting)

    sum_square_residues = n_st_sheets*(st_sheet[0]*st_sheet[1] - sheet[0]*sheet[1]*best_cutting)

    print(f"""Для листа {sheet[0]}x{sheet[1]}:
    Количество стандартных листов: {n_st_sheets}
    Суммарная площадь обрезков: {sum_square_residues}
    """)
```

__3. Решить задачу про курьеров, при наличии предпочтений относительно зон и часов у курьеров (описать формулу в общем случае)__

_Задача:_

Пусть есть компания, занимающаяся доставкой Belivery Club.
Есть N зон, в которых есть спрос на курьеров в каждый час. Внутри дня в разные часы спрос различен. Каждой зоне i в час h необходимо $n_{hi}$ курьеров. Для каждой зоны есть штраф из-за недостаточного 1 курьера - $s_i$

Необходимо распределить M курьеров таким образом, чтобы суммарный штраф был минимальный.

_Решение 1:_

Пронумеруем курьеров:
пусть $x_{ih}^i$ = 1, если k-курьер работает в час h в i зоне

Под предпочтениями зон и часов у курьеров будем понимать общие предпочтения у всех курьеров

Введем коэффициенты предпочтения для часов и зон, тогда:

$c_{i}$ - коэффициент предпочтения i-зоны

$t_{h}$ - коэффициент предпочтения h-часа

Тогда в одной зоне в час h работает $$\sum_{k=1}^{M} x_{ih}^k$$

Штраф в час в этой зоне
$$ f_{ih} = \begin{cases} 0, \space если \space \sum_{k=1}^{M} x_{ih}^k - n_{ih} \geq0 \\
s_i * (\sum_{k=1}^{M} x_{ih}^k - n_{ih}), в\spaceпротивном\spaceслучае. \end {cases}$$ 

Штраф за день $$\sum_{h=1}^{24}f_{ih} t_{h}$$

Штраф по всем зонам в течении дня:

$$\sum_{i=1}^{N} c_{i} \sum_{h=1}^{24}f_{ih} t_{h}$$

Целевая функция:

$$\sum_{i=1}^{N} c_{i} \sum_{h=1}^{24}f_{ih} t_{h} \rightarrow min $$

_Решение 2:_

Под предпочтениями зон и часов у курьеров будем понимать предпочтения часов и зон каждого курьера

Введем коэффициенты предпочтения для часов и зон, тогда:

$c_{i}^k$ - признак предпочтения i-зоны для k-курьера (1 - предпочитает, 0 - нет)

$t_{h}^k$ - признак предпочтения h-часа для k-курьера (1 - предпочитает, 0 - нет)

Тогда в зоне i в час h работает курьеров

$$\sum_{k=1}^{M} c_{i}^k t_{h}^k$$

Штраф в час в этой зоне
$$ f_{ih} = \begin{cases} 0, \space если \space \sum_{k=1}^{M} c_{i}^k t_{h}^k - n_{ih} \geq 0 \\
s_i * (\sum_{k=1}^{M} c_{i}^k t_{h}^k - n_{ih}), в\spaceпротивном\spaceслучае. \end {cases}$$ 

Штраф за день $$\sum_{h=1}^{24}f_{ih}$$

Штраф по всем зонам в течении дня:

$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih}$$

Целевая функция:

$$\sum_{i=1}^{N}\sum_{h=1}^{24}f_{ih} \rightarrow min $$
